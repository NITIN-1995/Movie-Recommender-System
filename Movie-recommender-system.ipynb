```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_credits.csv')
credits=pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
# merge both data set to single data of movies using title
movies=movies.merge(credits,on='title') 
```

```python
movies.info()
```

```python
#what column to keep for content based recommender system
movies=movies[['movie_id','cast','crew','genres','keywords','overview','title']]
```

```python
movies.head()
```

```python
#check missing data
movies.isnull().sum()
```

```python
#drop null values row
movies.dropna(inplace=True)
```

```python
#check for duplicate values
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#convert obj object from str to list
import ast

def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l

        
        
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#apply similar to keyword column
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast'][0]    # or movies.iloc[0].cast
```

```python
def convert3(obj):
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            l.append(l['name'])
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director' :
            l.append(i['name'])
            break
            
            
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# overviw column is str so we convert it to list by applying lamda function so that we can cocatenate it with other columns
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# we will reove space ex: space war--> spacewar , bcoz space can also be interpreted as space colon or other space releated words
movies['cast']=movies['cast'].apply(lambda x: [i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(' ','') for i in x])
movies['overview']=movies['overview'].apply(lambda x: [i.replace(' ','') for i in x])
movies['genres']=movies['genres'].apply(lambda x: [i.replace(' ','') for i in x])
            
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: ' '.join(x))
```

```python
new_df.head()
```

```python
#convert to lower case
new_df['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray() #converted every movie to vector
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    l=[]
    for i in text.split():      #split function will convert string to list and we will iterate over it to get each word
        l.append(ps.stem(i))
        
    return ' '.join(l)        #convert list to string
   
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
new_df[new_df['title']=='Batman Begins'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
new_df.iloc[1920].title
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0] # for index of movie
    distance = similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]   # lambda is used to get 1st element for sorting
    for i in movie_list:
        print(new_df.iloc[i[0]].title)          #i[0] is index of the movie
        
    
    
    
```

```python
recommend('Avatar')
```
